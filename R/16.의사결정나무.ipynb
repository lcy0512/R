```r
# install.packages('caret')
library(caret)
library(rpart)
# 학습용 검증용 분류
set.seed(1234)
samp <- createDataPartition(iris$Species, p = 0.7, list = F)
```

```r
# 학습용
data.tr <- iris[samp,]
# 검증용
data.test <- iris[-samp,]
```

```r
# 문제 및 정답 만들기
x <- subset(data.test, select = -Species)
y <- data.test$Species

nrow(x)
length(y)
```

```r
# 학습
iris_tree <- rpart(Species ~ ., data = data.tr)
```

```r
# 검증
pred <- predict(iris_tree, newdata = x, type = 'class')
pred
```

```r
# 예측력
mean(pred == y)
```

```r
# 오분류표
table(pred,y)
```
