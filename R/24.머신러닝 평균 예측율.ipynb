```r
# iris data를 randomForest로 예측율 구하기
rfFun <- function(count)  {
    library(randomForest)
    library(caret)
    result <- c()
    for(i in 1:count) {
        samp <- createDataPartition(iris$Species, p=0.7, list=F)
        result <- c(result,mean(predict(randomForest(Species ~ ., data = iris[samp,], ntree=100), subset(iris[-samp,], select = -Species)) == iris[-samp,]$Species)) # 예측력
    }
    cat("\033[45m------------------------------------------------\033[0m\n")
    cat("\033[36m 예측율 \033[0m\n",result,"\n")
    # cat("\n------------------------------------------------\n")
    cat("\033[45m------------------------------------------------\033[0m\n")
    cat("\033[31m 평균 예측율 \033[0m\n",mean(result),"\n")
    # cat("\n------------------------------------------------")
    cat("\033[45m------------------------------------------------\033[0m\n")
}
```

```r
rfFun(100)
```
