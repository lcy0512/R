```r
# 기본 옵션
options(warn = -1)
options(repr.plot.width=15, repr.plot.height=9)
options(scipen = 100)
Sys.setlocale(category = "LC_ALL", locale = "ko_KR.UTF-8")
```

```r
# 밀가루 사용
flour <- c(3, -2, -1, 0, 1, -2)
# 다이어트 약 사용
diet <- c(-4, 1, -3, -5, -2, -8)

total <- c(flour, diet)
```

```r
# 기본 통계량
summary(total)

# 히스토그램, boxplot
par(bg = 'white', mfrow = c(1,2))
hist(total)
boxplot(total)
```

```r
par(bg = 'white')
boxplot(flour,diet)

# 밀가루약이 효과가 없다는 것을 입증하려면 밀가루약과 다이어트약의 상관계수를 확인한다.
# 밀가루약과 다이어트약은 상관관계가 없다는 결론, 따라서 밀가루약은 약효가 없다.
cor(flour, diet)
```

# 기초 통계량 산출

```r
cafe <- read.csv("Data/cafe.csv")
```

```r
# 빈도수 : 범위가 너무 다양한 문제점이 있다.
table(cafe$Coffee)

# 범위 빈도수 : 내가 직접 범위를 지정하여 그 범위에 해당하는 빈도수를 구한다.
coffee_Range <- seq(0, 60, 10)
coffee_Range
table(cut(cafe$Coffee, breaks = coffee_Range)) # 0 < x <= 10 / 10 < x <= 20
table(cut(cafe$Coffee, breaks = coffee_Range, right = F)) # 0 <= x < 10 / 10 <= x < 20

# 줄기 - 잎 그림
stem(cafe$Coffee)
```

```r
# Coffee 판매량의 변동계수
sd(cafe$Coffee) / mean(cafe$Coffee)
# Juice 판매량의 변동계수
sd(cafe$Fruit.Cup) / mean(cafe$Fruit.Cup)
```

```r
sort(cafe$Fruit.Cup)
sort(cafe$Coffee)
```

```r
# boxplot으로 정규분포를 이루고 있는지를 확인
par(bg = 'white')
boxplot(
    cafe$Coffee,cafe$Fruit.Cup
)
```
